```python

# Filter the dataset for first-owner bikes
first_owner_bikes = df[df['owner'] == '1st owner']

# Plot a scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(x='km_driven', y='selling_price', data=first_owner_bikes)
plt.title('Relationship between Selling Price and Km Driven (1st Owner Bikes)')
plt.xlabel('Km Driven')
plt.ylabel('Selling Price (INR)')
plt.show()

# Calculate the correlation between selling_price and km_driven
correlation = first_owner_bikes['selling_price'].corr(first_owner_bikes['km_driven'])

# Print the correlation coefficient
print(f"The correlation between selling price and kilometers driven for first-owner bikes is {correlation:.2f}")
```

 Identify and remove outliers in the km_driven column using the IQR method

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for km_driven
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)

# Calculate IQR (Interquartile Range)
IQR = Q3 - Q1

# Define the lower and upper bounds for the acceptable range of km_driven
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the dataset to remove outliers in the km_driven column
df_no_outliers = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]

# Print the number of rows before and after removing outliers
print(f"Number of rows before removing outliers: {df.shape[0]}")
print(f"Number of rows after removing outliers: {df_no_outliers.shape[0]}")
```

Perform a bivariate analysis to visualize the relationship between year and selling_price


```python
# Scatter plot to visualize the relationship between year and selling_price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='year', y='selling_price', data=df)
plt.title('Scatter Plot: Year vs Selling Price')
plt.xlabel('Year of Manufacture')
plt.ylabel('Selling Price (INR)')
plt.show()

# Box plot to show the distribution of selling_price for each year
plt.figure(figsize=(12, 6))
sns.boxplot(x='year', y='selling_price', data=df)
plt.title('Box Plot: Year vs Selling Price')
plt.xlabel('Year of Manufacture')
plt.ylabel('Selling Price (INR)')
plt.xticks(rotation=45)
plt.show()
```

 What is the average depreciation in selling price based on the bike's age (current year - manufacturing
year)

```python
# Get the current year
current_year = 2024  # You can also use datetime to get the current year automatically

# Calculate the age of the bike (current year - manufacturing year)
df['age'] = current_year - df['year']

# Group by age and calculate the average selling price for each age
avg_selling_price_by_age = df.groupby('age')['selling_price'].mean().reset_index()

# Print the average selling price by age
print(avg_selling_price_by_age)


# Plotting the depreciation trend based on bike age
plt.figure(figsize=(10, 6))
sns.lineplot(x='age', y='selling_price', data=avg_selling_price_by_age)
plt.title('Average Depreciation in Selling Price Based on Bike Age')
plt.xlabel('Bike Age (years)')
plt.ylabel('Average Selling Price (INR)')
plt.show()
```

Which bike names are priced significantly above the average price for their manufacturing year

```python



# Calculate the average selling price by manufacturing year
avg_price_by_year = df.groupby('year')['selling_price'].mean().reset_index()
avg_price_by_year.rename(columns={'selling_price': 'avg_selling_price'}, inplace=True)

# Merge the average selling price by year with the original dataset
df_with_avg = pd.merge(df, avg_price_by_year, on='year')

# Define the threshold for "significantly above" (e.g., 20% above the average price)
threshold_percentage = 0.2
df_with_avg['significantly_above_avg'] = df_with_avg['selling_price'] > (df_with_avg['avg_selling_price'] * (1 + threshold_percentage))

# Filter the bikes that are significantly above the average price for their manufacturing year
bikes_significantly_above_avg = df_with_avg[df_with_avg['significantly_above_avg']]

# Display the bike names that are significantly above the average price
bikes_significantly_above_avg[['name', 'year', 'selling_price', 'avg_selling_price']]
```

Develop a correlation matrix for numeric columns and visualize it using a heatmap.

```python
# Select numeric columns for correlation matrix
numeric_cols = df.select_dtypes(include=['number']).columns

# Calculate the correlation matrix
corr_matrix = df[numeric_cols].corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Set title
plt.title('Correlation Matrix of Numeric Columns')

# Show the plot
plt.show()
```
