```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
## To check dataType
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
# Pie Chart  Top 5 country use Zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.1f%%')
```

```python
ratings=final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating  Count"})
```

```python
ratings
```

## Observations

1. Rating is Between 4.5 to 4.9 ----> Excellent
2. Rating is Between 4.0 to 4.4 ----> Very Good
3. Rating is Between 3.5 to 3.9 ----> Good
4. Rating is Between 2.5 to 3.4 ----> Average
5. Rating is Between 1.8 to 2.4 ----> Poor
6. Rating is 0                   ----> Not Rated

```python
ratings = final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size().reset_index(name="Rating Count")

import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['figure.figsize'] = (12, 6)

sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
plt.show()
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue = 'Rating color',data=ratings,palette=['white','red','yellow','green'])
```

```python
## Coiunt Plot
sns.countplot(x="Rating color",data=ratings,palette=['white','red','yellow','green'])
```
